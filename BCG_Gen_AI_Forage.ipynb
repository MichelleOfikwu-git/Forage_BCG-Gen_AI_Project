```python
import pandas as pd
```

```python
df = pd.read_csv('10-K filings for Microsoft Tesla & Apple.csv')
```

```python
df.head(10)
```

```python
df['Total Revenue'] = df['Total Revenue'].astype(str).str.replace(',', '', regex=False)
df['Total Revenue'] = pd.to_numeric(df['Total Revenue'], errors='coerce')
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100

display(df)
```

# Task
Analyze the provided financial data by applying different aggregate functions and groupings to explore trends and patterns across companies and over fiscal years. Specifically, group the data by 'Company' and 'Fiscal Year' and calculate relevant aggregate statistics for numerical columns. Additionally, explore other potential aggregations and groupings to gain further insights.

## Group by company and calculate aggregates

### Subtask:
Group the DataFrame by 'Company' and calculate the mean of the numerical columns ('Total Revenue', 'Revenue Growth (%)').


**Reasoning**:
Group the DataFrame by 'Company' and calculate the mean of 'Total Revenue' and 'Revenue Growth (%)'.



```python
df_agg = df.groupby('Company')[['Total Revenue', 'Revenue Growth (%)']].mean()
display(df_agg)
```

## Group by fiscal year and calculate aggregates

### Subtask:
Group the DataFrame by 'Fiscal Year' and calculate the mean of the numerical columns ('Total Revenue', 'Revenue Growth (%)').


**Reasoning**:
Group the DataFrame by 'Fiscal Year' and calculate the mean of the specified numerical columns.



```python
df_agg_year = df.groupby('Fiscal Year')[['Total Revenue', 'Revenue Growth (%)']].mean()
display(df_agg_year)
```

## Explore other aggregations

### Subtask:
Consider other aggregate functions like sum, median, min, max, or count, and apply them to relevant columns.


**Reasoning**:
The subtask requires calculating various aggregate statistics by grouping the data. I will group the DataFrame by 'Company' and calculate the sum of 'Total Revenue'. Then I will group by 'Fiscal Year' and calculate the median of 'Total Revenue'. After that, I will group by 'Company' and calculate the count of non-null 'Revenue Growth (%)'. Finally, I will group by both 'Company' and 'Fiscal Year' and calculate the maximum of 'Total Revenue' as an additional exploration.



```python
# Filter out the row where 'Company' is 'Company'
df_filtered = df[df['Company'] != 'Company'].copy()

df_company_sum_revenue = df_filtered.groupby('Company')['Total Revenue'].sum().reset_index()
display(df_company_sum_revenue)

df_year_median_revenue = df_filtered.groupby('Fiscal Year')['Total Revenue'].median().reset_index()
display(df_year_median_revenue)

df_company_count_growth = df_filtered.groupby('Company')['Revenue Growth (%)'].count().reset_index()
display(df_company_count_growth)

df_company_year_max_revenue = df_filtered.groupby(['Company', 'Fiscal Year'])['Total Revenue'].max().reset_index()
display(df_company_year_max_revenue)
```

## Analyze other columns

### Subtask:
Explore aggregating other numerical columns like 'Total Assets', 'Total Liabilities', and 'Cash Flow from Operating Activities' after cleaning them similarly to 'Total Revenue'.


**Reasoning**:
Convert the specified columns to string, remove non-numeric characters, convert to numeric, and then perform the requested aggregations by Company and Fiscal Year.



```python
cols_to_clean = ['Total Assets', 'Total Liabilities', 'Cash Flow from Operating Activities']

for col in cols_to_clean:
    df[col] = df[col].astype(str).str.replace(',', '', regex=False).str.replace('+', '', regex=False)
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Filter out the row where 'Company' is 'Company'
df_filtered = df[df['Company'] != 'Company'].copy()

df_agg_company = df_filtered.groupby('Company')[cols_to_clean].mean()
display(df_agg_company)

df_agg_year = df_filtered.groupby('Fiscal Year')[cols_to_clean].mean()
display(df_agg_year)
```
